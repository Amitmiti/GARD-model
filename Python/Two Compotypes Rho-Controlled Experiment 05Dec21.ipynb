```python
def smoothingFunc(org_vector, smoothing_bin_size):
    new_vector = []
    for counter_key,key in enumerate(org_vector):
        if counter_key < (smoothing_bin_size) / 2:
            new_val = np.mean(org_vector[:int(counter_key + ((smoothing_bin_size - 1)/2) + 1)])
            new_vector.append(new_val)
        elif counter_key > len(org_vector) - ((smoothing_bin_size) / 2):
            new_val = np.mean(org_vector[int(counter_key - ((smoothing_bin_size - 1)/2)):])
            new_vector.append(new_val)
        else:
            new_val = np.mean(org_vector[int(counter_key - ((smoothing_bin_size)/2)):int(counter_key+((smoothing_bin_size)/2)+1)])
            new_vector.append(new_val)
    return new_vector
```
